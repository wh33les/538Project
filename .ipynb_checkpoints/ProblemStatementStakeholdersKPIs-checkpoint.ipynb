# Problem statment, stakeholders, and KPIs

## First look at the data 

There are two data files to look at.

```python
# Import pandas to load the data
import pandas as pd

# Load the first data set
raw_data1 = pd.read_csv("data/ocean157ksafe.csv")

# Show the data
raw_data1
```

```python
# Columns for the first data set
raw_data1.columns.values.tolist()
```

The column descriptions are given on the [Kaggle](https://www.kaggle.com/datasets/thedevastator/mental-health-in-drug-users-during-covid-19) page.  A few columns are in Spanish but we can translate them manually using Google translate.  Then we can create a new data file with the translated columns.

```python
# Create the data frame that will have the translated columns
translated_data1 = raw_data1

# There are only a few columns in the first data file that need translating
translated_data1.rename(columns={"2_edad":"2_age", 
                      "2_genero":"2_gender",
                      "2_nacionalidad":"2_nationality",
                      "2_pais_actual":"2_current_country"},
            inplace=True)

# Save the new data frame as a .csv file
translated_data1.to_csv("data/translated_data1.csv")
```

Now we can look at the second data file. 

```python
# Load the second data set
raw_data2 = pd.read_csv("data/concienciaysustancia10k7safe.csv")

# Show the data
raw_data2
```

```python
# Columns for the second data set
raw_data2.columns.values.tolist()
```

In this case most of the columns are in Spanish.  However, there is a translator module we can use in a loop to translate the columns all at once.    

```python
# Import the Translator class and assign it to a variable
from googletrans import Translator
translator = Translator()

# Create another dataframe with the same data as raw_data2 that will have the columns translated to English
translated_data2 = raw_data2

# Create a loop that will translate each column name
for spanish_name in raw_data2.columns.values.tolist():
    to_english = translator.translate(spanish_name)
    translated_data2.rename(columns={spanish_name:to_english.text}, # to_english.text is a list that includes the translation 
                 inplace=True) # inplace ignores the copy of the data frame created

# Show the translated columns
translated_data2.columns.values.tolist()            
```

The translator isn't perfect but we can change the untranslated columns manually.

```python
# Manually translate the untranslated column names
translated_data2.rename(columns={"2_programa_microdosificacion":"2_microdosing_program", 
                      "6_psychoactive_cosumidos":"6_psychoactives_consumed",
                      "9_feliz":"9_happy",
                      "11_encaro_oblicaciones_sin_problemas":"11_face_obligations_without_problems"},
            inplace=True)

# Show the modified columns
translated_data2.columns.values.tolist()
```

Finally, save the translated data frame as a .csv file.

```python
translated_data2.to_csv("data/translated_data2.csv")
```

## Problem statement

Is there a correlation between the use of drugs and overall mental health during the COVID pandemic?

## Stakeholders

This data analysis will be of interest to policy makers who legislate drug use.  Mental health professionals will be interested in the results when finding new treatments.

## Key performance indicators (KPIs)

The five personality traits measured in the first data set depend on the amount of drugs used.  The results from the second data set can be tallied into an overall mental health score that can be visualized as a function of the amount of drugs used.
