The second data set includes the drug use information and more specific mental health indicators for all the respondents.  The column names for the second data set are all in Spanish.  

```python
# Load the second data set
raw_data2 = pd.read_csv("concienciaysustancia10k7safe.csv")

# Show the data
raw_data2
```

We can use a translator to rename the columns in English.  

```python
# Import the Translator class and assign it to a variable
from googletrans import Translator
translator = Translator()

# Create another dataframe with the same data as raw_data2 that will have the columns translated to English
data2 = raw_data2

# Create a loop that will translate each column name
for spanish_name in data2.columns.values.tolist():
    to_english = translator.translate(spanish_name)
    data2.rename(columns={spanish_name:to_english.text}, # to_english.text is a list that includes the translation 
                 inplace=True) # inplace ignores the copy of the data frame created

# Show the translated columns
data2.columns.values.tolist()            
```

The translator isn't perfect but we can change the untranslated columns manually.

```python
# Manually translate the untranslated column names
data2.rename(columns={"2_programa_microdosificacion":"2_microdosing_program", 
                      "6_psychoactive_cosumidos":"6_psychoactives_consumed",
                      "9_feliz":"happy",
                      "11_encaro_oblicaciones_sin_problemas":"11_face_obligations_without_problems"},
            inplace=True)

# Show the modified columns
data2.columns.values.tolist()
```

## Problem statement

Is there a correlation between the use of drugs and overall mental health during the COVID pandemic?

## Stakeholders

This data analysis will be of interest to policy makers who legislate drug use.  Mental health professionals will be interested in the results when finding new treatments.

## Key performance indicators (KPIs)

The five personality traits measured in the first data set depend on the amount of drugs used.  The results from the second data set can be tallied into an overall mental health score that can be visualized as a function of the amount of drugs used.
